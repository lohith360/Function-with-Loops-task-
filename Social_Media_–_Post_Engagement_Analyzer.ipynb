```python
def analyze_engagement(likes_list):
    total = sum(likes_list)
    status = "Viral Post" if total >= 1000 else "Normal Engagement"
    return total, status

likes = [int(x) for x in input("Enter likes per day separated by space: ").split()]
t, s = analyze_engagement(likes)
print(f"Total Likes: {t}\nPost Status: {s}")
```
