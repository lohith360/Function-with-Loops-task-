```python
def check_rainfall(data, required):
    avg = sum(data) / len(data) if data else 0
    status = "Adequate Rainfall" if avg >= required else "Inadequate Rainfall"
    return avg, status

rain = [float(x) for x in input("Enter rainfall data separated by space: ").split()]
req = float(input("Enter required average rainfall: "))
a, s = check_rainfall(rain, req)
print(f"Average Rainfall: {a}\nRainfall Status: {s}")
```
