```python
def check_attendance(record):
    perc = (sum(record) / len(record)) * 100 if record else 0
    status = "Eligible" if perc >= 75 else "Not Eligible"
    return perc, status

record = [int(x) for x in input("Enter 1 (present) or 0 (absent) separated by space: ").split()]
p, s = check_attendance(record)
print(f"Attendance Percentage: {p}\nExam Eligibility: {s}")
```
